```python
#load libary pandas
import pandas as pd
```

```python
# read csv file(reg_pbp_2009)
data = pd.read_csv("reg_pbp_2009.csv")
data
```

```python
#get the shape
data.shape
```

```python
#According to the percentage NAN in each column
def percnt_missing(data):
    percnt_missing_data = data.isna().sum() / data.shape[0]
    return percnt_missing_data
```

```python
#numbers of columns have missing data
def colunm_missing(data):
    percnt_missing_data = percnt_missing(data)
    count=0
    for index in data:
        if(percnt_missing_data[index]>0):
            count+=1
    print("The number of columns that contain NAN :",count)
```

```python
percnt_missing(data)
```

```python
colunm_missing(data)
```

```python
# Delete columns containing either 70% or more than 70% NaN Values
def clean(data,percnt_missing):
    min_count =  int(((100-percnt_missing)/100)*data.shape[0] + 1)
    data = data.dropna( axis=1, thresh=min_count)
    return data
```

```python
data = clean(data,70)
```

```python
colunm_missing(data)
```

```python
#fill missing values by mean
data=data.fillna(data.mean()).round(2)
```

```python
colunm_missing(data)
```

```python
#to select all columns with NaN values:
data[data.columns[data.isna().any()]]
```
